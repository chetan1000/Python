```python
# def is Keyword in python
```

```python
# creating a Function
def first_func():                # func name or func definition
    print('We did it!')         # func body
    
    
first_func()            # func call
```

```python
# Passing Argument into Function

def number_squared(number):
    print(number ** 2)
    

number_squared(5)
```

```python
# Passing Two or multiple arguments into function

def number_squared_cust(number, power):
    print(number ** power)
    
     
number_squared_cust(5, 3)                       # automatically 5 is assigned to number, and 3 is to power
```

### Arbitary argument

```python
# if we don't know how many arguments are passing
```

```python
def number_args(*number):
    print(number[0] * number[1] * number[2])
    
    
number_args(1,2,3,4,5,6)            # we are passing many arguments as Tuple
```

```python
args_tuple = (5, 6, 1, 2, 8)

def number_args(*number):
    print(number[0] * number[1] * number[2]) 
    

    
# number_args(args_tuple)                                # IndexError: tuple index out of range
    
number_args(*args_tuple)
```

```python
def number_args(*number):
    print(number[0] * number[1] * number[2])
    
    
number_args(*[1,2,3,4,5,6])

# number_args(*(1,2,3,4,5,6))
```

### KeyWord Argument

```python
def number_squared_cust(number, power):
    print( number ** power)
    
    
number_squared_cust(power = 5, number = 3)
```

### Arbitary KeyWord Argument

```python
# We don't know , how many Keyword arguents are passing in func definition
```

```python
def number_kwarg(**number):
    print('My number is: ' + number['integer'] + 'My other number: ' + number['integer2'])
    
    
    

number_kwarg(integer= '2309', integer2='349')                  # keword assignment
                                                                # '2309', '349' are string
```
