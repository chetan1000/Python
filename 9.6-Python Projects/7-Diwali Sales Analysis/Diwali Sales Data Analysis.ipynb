# Exploratory Data Analysis

## Gender

```python
df.columns
```

```python
ax = sns.countplot(x  = 'Gender', data = df)
```

```python
ax = sns.countplot(x  = 'Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)
```

```python
sales_Gender_wise = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending = False)

ay = sns.barplot(x = 'Gender', y='Amount', data = sales_Gender_wise)

for bars in ay.containers:
    ay.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and purchasing power of females is greater than man

## Age

```python
df.columns
```

```python
sns.countplot(data = df, x='Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graph, we can say that 26-35 age range Females did shoping more

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_age
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(x='Age Group',y='Amount', data = sales_age)


ax.bar_label(ax.containers[0], fmt='%.0f', color='black', fontsize=8)

# Show the plot
plt.show()
```

From the above graph we can see that most of buyers or spenders are of 26-35 age group peoples

## State

```python
df.columns
```

```python
# total number of orders from top 10 states
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sales_state
```

```python
sales_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16, 5)})
ax= sns.barplot(data = sales_state, x='State',y='Orders')
ax.bar_label(ax.containers[0], fmt='%.0f', color='black', fontsize=10)
plt.show()
```

```python
#total amount/sales of sales from top 10 states

sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sns.set(rc={'figure.figsize':(16, 5)})
ax= sns.barplot(data = sales_state, x='State',y='Amount')
ax.bar_label(ax.containers[0], fmt='%.0f', color='black', fontsize=10)
plt.show()
```

From above graphs most of the orders are from UP, Maharastra, Karnataka

## Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(7,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)

sns.set(rc={'figure.figsize':(6,6)})
ay = sns.barplot(data = sales_state, x='Marital_Status', y='Amount', hue ="Gender")


for container in ay.containers:
    ay.bar_label(container, fmt='%.0f', color='black', fontsize=10)
```

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x= 'Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.set(rc={'figure.figsize':(20,5)})

az = sns.barplot(data = sales_state, x = 'Occupation', y = 'Amount')

for container in az.containers:
    az.bar_label(container, fmt='%.0f', color='black', fontsize=12, fontweight='bold')
```

From above graph, we can see that most of the buyers are working at IT Sectors, HealthCare and Aviation

## Product Category

```python
sns.set(rc={'figure.figsize':(25,10)})
am = sns.countplot(data = df, x = 'Product_Category')

for bars in am.containers:
    am.bar_label(bars, fmt='%.0f', color='black', fontsize=18, fontweight='bold')
    
    am.set_xticklabels(am.get_xticklabels(), fontsize=15,fontweight='bold')
    


# Rotate x-axis labels for better visibility
    plt.xticks(rotation=45, ha='right')  # Rotate labels and align them to the right

    plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
    plt.show()
```

 People spent most of the money for clothing & Apparel and then Food and Electronics 

```python
sales_state = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(data=sales_state,x='Product_Category', y='Amount')


az = sns.barplot(data = sales_state, x = 'Product_Category', y = 'Amount')

for container in az.containers:
    az.bar_label(container, fmt='%.0f', color='black', fontsize=12, fontweight='bold')
    
    

plt.xticks(rotation=45, ha='right')  # Rotate labels and align them to the right
az.set_xticklabels(az.get_xticklabels(), fontsize=15,fontweight='bold')

plt.tight_layout() 
plt.show()
```

From above graph, most of the amount spent on foods, but if you see the graph, which is above then this graph Clothing has highest oder

Most sold proucts are Food , Clothes and Electronics

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sns.set(rc={'figure.figsize':(20, 7)})
sns.barplot(data = sales_state, x='Product_ID',y='Orders')

az = sns.barplot(data = sales_state, x = 'Product_ID', y = 'Orders')

for container in az.containers:
    az.bar_label(container, fmt='%.0f', color='black', fontsize=12, fontweight='bold')
    
az.set_xticklabels(az.get_xticklabels(), fontsize=15,fontweight='bold')
plt.show()
```
